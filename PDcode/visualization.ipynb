```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/alessandrocantarelli/Documents/Tec/PD sensor extracted data/main_sleep/15-16/sleepCN.csv')
data = data.dropna(subset=['arm_angle'])
data_1 = pd.read_csv('/Users/alessandrocantarelli/Documents/Tec/PD sensor extracted data/main_sleep/16-17/sleepCN.csv')
data_1 = data_1.dropna(subset=['arm_angle'])
data_2 = pd.read_csv('/Users/alessandrocantarelli/Documents/Tec/PD sensor extracted data/main_sleep/17-18/sleepCN.csv')
data_2 = data_2.dropna(subset=['arm_angle'])
```

```python
## combine these different collections into a list
data_to_plot = [data['arm_angle'],data_1['arm_angle'], data_2['arm_angle']]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1.5,1.5])

# Create the boxplot
bp = ax.violinplot(data_to_plot, showmeans=True, showextrema=True)

bp['bodies'][0].set_facecolor('red')
bp['bodies'][1].set_facecolor('green')
bp['bodies'][2].set_facecolor('yellow')
bp['bodies'][0].set_edgecolor('black')
bp['bodies'][1].set_edgecolor('black')
bp['bodies'][2].set_edgecolor('black')
bp['bodies'][0].set_alpha(0.5)
bp['bodies'][1].set_alpha(0.5)
bp['bodies'][2].set_alpha(0.5)
    
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Sleep Episode 1', 'Sleep Episode 2', 'Sleep Episode 3'],  fontsize=20)
plt.ylabel('Arm Angle',  fontsize=20)
plt.title('Distribution of Arm Angle during Sleep: Control Participant 1011', fontsize=20)
plt.show()
```
