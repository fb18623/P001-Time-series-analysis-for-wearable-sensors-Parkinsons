```python
def separate_participants(data, label):
    grouped = data.groupby(data.id)
    df1 = grouped.get_group(label)
    df1 = df1.sort_values(by=['bt'])
    df1 = df1.reset_index(drop=True)
    return df1
```

```python
def separate_data(data):
    timestamp = []
    x = []
    y = []
    z = []
    for i in range(len(data)):
        test = data.iloc[i]
        # getting correct timestamp
        date = pd.to_datetime(test[4])
        t1 = dt.timedelta(seconds=test[21])
        t2 = dt.timedelta(seconds=test[17])
        t3 = dt.timedelta(seconds=test[13])
        t4 = dt.timedelta(seconds=test[9])
        t5 = dt.timedelta(seconds=test[5])
        timestamp.append(date)
        timestamp.append(date + t1)
        timestamp.append(date + t2)
        timestamp.append(date + t3)
        timestamp.append(date + t4)
        timestamp.append(date + t5)
        # extract x,y,z
        x.append(test[26])
        y.append(test[27])
        z.append(test[28])
        x.append(test[22])
        y.append(test[23])
        z.append(test[24])
        x.append(test[18])
        y.append(test[19])
        z.append(test[20])
        x.append(test[14])
        y.append(test[15])
        z.append(test[16])
        x.append(test[10])
        y.append(test[11])
        z.append(test[12])
        x.append(test[6])
        y.append(test[7])
        z.append(test[8])
    dataframe = pd.DataFrame({"t": timestamp, "x": x, "y": y, "z": z})
    return dataframe
```
